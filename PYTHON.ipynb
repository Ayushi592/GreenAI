```python
def calculator():
    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))
    op = input("Choose operation (+, -, *, /): ")

    if op == '+':
        print(num1 + num2)
    elif op == '-':
        print(num1 - num2)
    elif op == '*':
        print(num1 * num2)
    elif op == '/':
        print(num1 / num2 if num2 != 0 else "Error! Division by zero.")
    else:
        print("Invalid operation!")

calculator()
```

```python
climate_data = [
    {"city": "City A", "temperature": 25, "carbon_footprint": 500},
    {"city": "City B", "temperature": 30, "carbon_footprint": 350},
    {"city": "City C", "temperature": 22, "carbon_footprint": 600},
    {"city": "City D", "temperature": 15, "carbon_footprint": 200},
    {"city": "City E", "temperature": 28, "carbon_footprint": 450},
]

# Filter cities with temperature above 26 degrees
high_temp_cities = [city["city"] for city in climate_data if city["temperature"] > 26]

print("Cities with temperature above 26 degrees:", high_temp_cities)
```

```python
total_carbon_footprint = sum(city["carbon_footprint"] for city in climate_data)
average_carbon_footprint = total_carbon_footprint / len(climate_data)
print("Average carbon footprint:", average_carbon_footprint)
```

```python
highest_carbon_city = max(climate_data, key=lambda city: city["carbon_footprint"])
print("City with the highest carbon footprint:", highest_carbon_city["city"])
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Solar" : [1200,1500,1300],
    "Wind": [3400,3600,3200],
    "Hydropower": [2900,3100,2800],
    "Biomass" : [2500,2700,2400]
}
df= pd.DataFrame(data)
sns.pairplot(df)
plt.show()
```

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidth=0.5)
plt.title('Correlation Between Energy Sources')
plt.show()
```

```python
sns.boxplot(data=df)
plt.title('Distribution of Energy Consumption by Source')
plt.show()
```

```python
sns.violinplot(data=df)
plt.title('Violin Plot of Energy Consumption Distributions')
plt.show()
#explore iris flower database
```
